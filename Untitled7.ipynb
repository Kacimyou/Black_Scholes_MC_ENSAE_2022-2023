```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
!pip install openpyxl
!pip install xlrd
ocde = pd.read_excel('/content/KEI_27102022043534511 (1).xlsx')
glob = pd.read_excel("/content/Global Real Econ Activity index (2).xlsx")
gscpi = pd.read_excel("/content/gscpi_data (3).xls", sheet_name = 1) # pour ce fichier il va falloir modifier les dates, on ouvre la deuxième feuille
#enlever les variables qui ne nous intéressent pas #
ocde = ocde.drop(["SUBJECT", "LOCATION", "MEASURE", "FREQUENCY", "TIME", "Unit Code", "PowerCode Code"], axis = 1)
ocde = ocde.drop(["Flag Codes", "Reference Period Code"], axis = 1)

ocde.info()
```

```python
list = ["Overnight interbank rate", "Long-term interest rate", "3 month interbank rate", "Broad money, s.a.", "Hourly earnings: manufacturing, s.a.", "Narrow money, s.a.", "Permits issued for dwellings, s.a.", "Total employment: all persons, s.a."]
for i in list:
    ocde = ocde[ocde["Subject2"] != i]


ocde["Subject2"].describe()
#convertir le format des dates#
ocde["Time"] = pd.to_datetime(ocde["Time"])
ocde.info()
ocde.head()
```

```python
ocde_USA = ocde[ ocde["Country"] == "United States"]
ocde_China = ocde[ ocde["Country"] == "China (People's Republic of)"]
ocde_euro = ocde[ ocde["Country"] == "Euro area (19 countries)"]
```

```python
ocde2 = ocde.pivot(index = "Time", columns = ["Country", "Subject2"], values = "Value")
ocde2.head()

```

```python
ocde_USA2 = ocde_USA.pivot(index= "Time", columns = "Subject2", values = "Value")
ocde_China2 = ocde_China.pivot(index= "Time", columns = "Subject2", values = "Value")
ocde_euro2 = ocde_euro.pivot(index= "Time", columns = "Subject2", values = "Value")
ocde_USA2.info()
```

```python
glob = glob.set_index('Date')
glob.head()
```

```python
merg_USA = ocde_USA2.merge(glob, how = "outer", left_index = True, right_index = True)
merg_USA.rename(columns = {"Unnamed: 1" : "Global real economic activity index"}, inplace = True)

merg_China = ocde_China2.merge(glob, how = "outer", left_index = True, right_index = True)
merg_China.rename(columns = {"Unnamed: 1" : "Global real economic activity index"}, inplace = True)

merg_euro = ocde_euro2.merge(glob, how = "outer", left_index = True, right_index = True)
merg_euro.rename(columns = {"Unnamed: 1" : "Global real economic activity index"}, inplace = True)
```

```python
gscpi2 = gscpi[4:] #enlève les lignes qui ne sont pas des données
gscpi2 = gscpi2[["Date", "GSCPI"]] #enlève lignes vides
```

```python
gscpi2["Date"] = pd.to_datetime(gscpi2["Date"])
              
gscpi2.info()
gscpi2.head(200)
```

```python
gscpi2['year'] = pd.DatetimeIndex(gscpi2['Date']).year
gscpi2['month'] = pd.DatetimeIndex(gscpi2['Date']).month
gscpi2['Date'] = pd.to_datetime(gscpi2[['year', 'month']].assign(DAY=1))

gscpi3 = gscpi2[["Date", "GSCPI"]]
gscpi3 = gscpi3.set_index('Date')
gscpi3.head()
```

```python
from pickle import NONE
merged_USA = merg_USA.merge(gscpi3, how = "outer", left_index = True, right_index = True)
merg_China = ocde_China2.merge(glob, how = "outer", left_index = True, right_index = True)
merg_China.rename(columns = {"Unnamed: 1" : "Global real economic activity index"}, inplace = True)

merg_euro = ocde_euro2.merge(glob, how = "outer", left_index = True, right_index = True)
merg_euro.rename(columns = {"Unnamed: 1" : "Global real economic activity index"}, inplace = True)

merged_China = merg_China.merge(gscpi3, how = "outer", left_index = True, right_index = True)
merged_euro = merg_euro.merge(gscpi3, how = "outer", left_index = True, right_index = True)
```

```python
def plot_comparatif(indicateur):
    plt.plot(merged_USA[indicateur], label = "USA")
    plt.plot(merged_China[indicateur], label = "Chine")
    plt.plot(merged_euro[indicateur], label = "Zone Euro")
    plt.legend()
    plt.title(indicateur)
    plt.show()

plot_comparatif("GDP (ratio to trend, smoothed)")
```
